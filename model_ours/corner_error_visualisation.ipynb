## Without Layer 1

```python
f = open('vanilla_layer2_corner_errors.txt', 'r')
errors = list(map(lambda x:float(x), f.readlines()))
```

```python
x = np.arange(41)
y = []
num_pts = len(errors)
for v in x:
    num = 0
    for e in errors:
        if e <= v:
            num += 1
    y.append(float(num * 100 / num_pts) )
y = np.asarray(y)
```

```python
fig, ax = plt.subplots(1)
plt.xlabel('Average Corner Error')
plt.ylabel('Percentage %')
ax.plot(x,y)
ax.set_title('Vanilla Initialisation')
```

```python
np.mean(errors)
```
